```python
import pandas as pd
import yahoo_fin.stock_info as si
```

```python
ticker_df = pd.read_csv('../ProjectDatasets/final_recommendations.csv', sep=',', index_col='ticker')
```

```python
tickersDF = si.tickers_sp500(True)
```

```python
tickersDF.index=tickersDF['Symbol']
tickersDF.index.rename('ticker', inplace=True)
tickersDF.drop(columns=['Symbol'], inplace=True)
```

```python
final=pd.concat([tickersDF,ticker_df], axis=1, join='inner')
```
